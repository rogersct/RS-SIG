```python
import timeit
import omap
import evr
```

```python
starttime = timeit.default_timer()
input_folder_path = './Train 23 14NOV2019 Work Folder - Copy/raw logs'
start_time = '2019/11/14 17:00:00' 
end_time = '2019/11/14 18:00:00'
# sample_output_filename = ['./Train 23 14NOV2019 Work Folder/preprocessing_output/OMAP_Train_23_Car_45_20191114_1700_to_20191114_1800.csv',
#                           './Train 23 14NOV2019 Work Folder/preprocessing_output/OMAP_Train_23_Car_46_20191114_1700_to_20191114_1800.csv']
omap_sample_output_filename = ['./Train 23 14NOV2019 Work Folder/preprocessing_output/OMAP_Train_23_Car_45_20191114_1700_to_20191114_1800.csv']
evr_sample_output_filename = ['./Train 23 14NOV2019 Work Folder/preprocessing_output/EVR_20191114_1700_to_20191114_1800.csv']

#Process OMAP
result_list = omap.process_omap(input_folder_path, start_time, end_time)
print("Time Taken:", timeit.default_timer() - starttime)
# Unit Test
for index in range (0, len(result_list)):
    omap.unit_test(omap_sample_output_filename[index], 'ATP', result_list[index][0])
    omap.unit_test(omap_sample_output_filename[index], 'COM', result_list[index][1])
    omap.unit_test(omap_sample_output_filename[index], 'TDMS', result_list[index][2])
    omap.unit_test_all(omap_sample_output_filename[index], result_list[index][3])


# Process EVR
df_c2_clean, df_od_clean, df_result = evr.process_evr(input_folder_path, start_time, end_time)
print("Time Taken:", timeit.default_timer() - starttime)
# Unit Test
#evr.output(sample_output_filename[0], df_result)
evr.unit_test(evr_sample_output_filename[0], 'C2', df_c2_clean)  
evr.unit_test(evr_sample_output_filename[0], 'OD', df_od_clean)  
evr.unit_test_all(evr_sample_output_filename[0], df_result)
```
